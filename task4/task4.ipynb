## Converting Alphabetic values  into Numerics

```python
label = LabelEncoder()
```

```python
iris_df['Species']=label.fit_transform(iris_df['Species'])
```

```python
iris_df
```

## Dependent And Independent Featurs

```python
X = iris_df.iloc[:,[1,2,3,4]].values
y = iris_df.iloc[:,-1].values
```

## Splitting into Training Set and Test Set

```python
X_train, X_test, y_train, y_test = train_test_split(
...     X, y, test_size=0.25, random_state=42)
```

```python
y_train
```

## Training Of Model

```python
clf = DecisionTreeClassifier(max_depth = 2, random_state = 0)
clf.fit(X_train, y_train)
clf.predict(X_test)
```

## Plots a decision tree using scikit-learn

```python
tree.plot_tree(clf)
```

```python
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa', 'versicolor', 'virginica']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
tree.plot_tree(clf,feature_names = fn, class_names=cn,filled = True)
```

## To Visualize the Decision Tree we have to  export Dot file

```python
tree.export_graphviz(clf,out_file="tree.dot",feature_names = fn,class_names=cn,filled = True)
```
